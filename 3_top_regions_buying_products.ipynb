```python
#import libarys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("clean_data_superstore_sales.csv")#reading the csv file
```

```python
#grouping the data according to the city and its Quantity sold also showing the top 10 cities
df_top_city=df.groupby("City")["Quantity"].sum().sort_values(ascending=False).head(10).to_frame()
```

```python
#making a costom_palette
costom_palette=sns.dark_palette("#69d", reverse=True, as_cmap=True)

#seting the style of the plot
sns.set_style("white")

#ploting the graph using seaborn
sns.barplot(df_top_city,x="Quantity",y=df_top_city.index,hue="Quantity",palette=costom_palette)

#Remove the top and right spines from plot
sns.despine()

# Get the current Axes object for further use
ax=plt.gca()

# Annotate each bar with its quantity value for better readability
for index,val in enumerate(df_top_city.Quantity):
    ax.text(val+70,index,val,va='center')

#removing the legend
plt.legend().set_visible(False)

#giving the axis label and graph title
plt.xlabel("Quantity sold",fontsize=13)
plt.ylabel("City",fontsize=13)
plt.title("Top Performing Cities Based on Quantity Sold",fontsize=16,weight="bold")

#showing the plot
plt.show()
```
